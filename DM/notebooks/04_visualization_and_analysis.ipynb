```python
# Now that we have all the data loaded from clustering, lets do some visualization and eval!

# for saving clusters:
BASE_FIG_DIR = Path("/home/jeezu/Meteorite-Data-Mining-Analysis/DM/outputs/figures/cluster plots")

# highest quality kmeans runs
display(
    summary.sort_values("best_silhouette", ascending=False)[
        ["run", "best_k", "best_silhouette", "n_rows_clustered"]
    ]
)

# our generic cluster plotting function. Now we cluster!
def plot_clusters_scatter(dfc, x_col, y_col, title, subdir="map"):
    out_dir = BASE_FIG_DIR / subdir
    out_dir.mkdir(parents=True, exist_ok=True)

    plt.figure(figsize=(8, 5))
    sns.scatterplot(
        data=dfc, x=x_col, y=y_col,
        hue="cluster", palette="tab10", s=12, alpha=0.7, linewidth=0
    )
    plt.title(title)
    plt.tight_layout()

    filename = f"{title.lower().replace(' ', '_')}_{x_col}_vs_{y_col}.png"
    out_path = out_dir / filename
    plt.savefig(out_path, dpi=220, bbox_inches="tight")
    plt.show()
    plt.close()

    print(f"Saved: {out_path}")


# display the centroid tables (averages for each value for each cluster)
display(main_centers)
display(fell_centers)
display(found_centers)
display(modern_centers)
display(historic_centers)

# DISPLAY CLUSTERS! AND IT WORKS WOOOOOOOOAH
# MAP: longitude vs latitude
plot_clusters_scatter(main, "reclong_norm", "reclat", "Main Clusters", subdir="map")
plot_clusters_scatter(main_with_fall, "reclong_norm", "reclat", "Main With Fall Clusters", subdir="map")
plot_clusters_scatter(fell, "reclong_norm", "reclat", "Fell Clusters", subdir="map")
plot_clusters_scatter(found, "reclong_norm", "reclat", "Found Clusters", subdir="map")
plot_clusters_scatter(modern, "reclong_norm", "reclat", "Modern Clusters", subdir="map")
plot_clusters_scatter(historic, "reclong_norm", "reclat", "Historic Clusters", subdir="map")

# TIME-MASS: year vs log mass
plot_clusters_scatter(main, "year", "log_mass", "Main Time-Mass Clusters", subdir="time_mass")
plot_clusters_scatter(main_with_fall, "year", "log_mass", "Main With Fall Time-Mass Clusters", subdir="time_mass")
plot_clusters_scatter(fell, "year", "log_mass", "Fell Time-Mass Clusters", subdir="time_mass")
plot_clusters_scatter(found, "year", "log_mass", "Found Time-Mass Clusters", subdir="time_mass")
plot_clusters_scatter(modern, "year", "log_mass", "Modern Time-Mass Clusters", subdir="time_mass")
plot_clusters_scatter(historic, "year", "log_mass", "Historic Time-Mass Clusters", subdir="time_mass")


```
